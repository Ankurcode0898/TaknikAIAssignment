```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
# Task 1: Create a NumPy array
names = ["Aarav","Priya","Rahul","Sneha","Vikram"]
```

```python
names
```

```python
marks = np.array([[78,85,90],[88,79,92],[70,65,80],[95,92,88],[60,70,75]])
marks
```

```python
# Task 2: Compute totals & averages

# (a) Total marks per student
total_marks = np.sum(marks,axis =1)
print("Total marks per student :",total_marks)

# (b) Average marks per subject
average_subject = np.mean(marks,axis=0)
print("Average marks per subject:",average_subject)

# (b) Highest m,arks in the whole class
highest_marks = np.max(marks)
print("Highest marks in whole class:",highest_marks)
```

```python
# Task 3: Convert to a Pandas DataFrame
names = ["Aarav","Priya","Rahul","Sneha","Vikram"]
import pandas as pd
df = pd.DataFrame(
    marks,
    columns=['Math', 'Science', 'English'],
    index=names
)

print(df.head)
print(df.describe())
```

```python
#Task 4: Filter top performers

df['total'] = df.sum(axis=1)
top_performer = df[df['total'] > 240]
print(top_performer[['total']])

```

```python
#Task 5:  Plot a bar chart
import matplotlib.pyplot as plt
plt.bar(df.index,df['total'])

plt.title("Total marks of students")
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.show()
```
